# “Batch Normalization - Improving Training Stability and Convergence in CNNs (CIFAR-10)”

```python
# Import Libraries

import numpy as np
import matplotlib.pyplot as plt

from tensorflow.keras.datasets import cifar10
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten
from tensorflow.keras.layers import Dense, BatchNormalization, Dropout
from tensorflow.keras.utils import to_categorical

plt.style.use("seaborn-v0_8-colorblind")
```

### Load the dataset

```python
(X_train, y_train), (X_test, y_test) = cifar10.load_data()

X_train = X_train / 255.0
X_test = X_test / 255.0

y_train = to_categorical(y_train, 10)
y_test = to_categorical(y_test, 10)

print("Train shape:", X_train.shape)
```

### Build a Model

```python
def create_cnn(use_bn=False, use_dropout=False):
    model = Sequential()

    model.add(Conv2D(32, (3,3), activation='relu', input_shape=(32,32,3)))
    if use_bn:
        model.add(BatchNormalization())

    model.add(MaxPooling2D())

    model.add(Conv2D(64, (3,3), activation='relu'))
    if use_bn:
        model.add(BatchNormalization())

    model.add(MaxPooling2D())

    model.add(Flatten())
    model.add(Dense(128, activation='relu'))

    if use_dropout:
        model.add(Dropout(0.5))

    model.add(Dense(10, activation='softmax'))

    model.compile(
        optimizer='adam',
        loss='categorical_crossentropy',
        metrics=['accuracy']
    )

    return model
```

### Train Models

```python
# Model 1: No BN
model_no_bn = create_cnn(use_bn=False)
history_no_bn = model_no_bn.fit(
    X_train, y_train,
    validation_split=0.2,
    epochs=8,
    batch_size=128,
    verbose=1
)

# Model 2: With BN
model_bn = create_cnn(use_bn=True)
history_bn = model_bn.fit(
    X_train, y_train,
    validation_split=0.2,
    epochs=8,
    batch_size=128,
    verbose=1
)
```

### Accuracy Plot

```python
plt.figure(figsize=(8,5))

plt.plot(history_no_bn.history['val_accuracy'], label='No BN')
plt.plot(history_bn.history['val_accuracy'], linestyle='--', label='With BN')

plt.xlabel("Epochs")
plt.ylabel("Validation Accuracy")
plt.title("Figure 1 – BN vs No BN Accuracy")
plt.legend()

plt.savefig("figure1_bn_accuracy.png", dpi=300)
plt.show()
```

### Loss Plot

```python
plt.figure(figsize=(8,5))

plt.plot(history_no_bn.history['val_loss'], label='No BN')
plt.plot(history_bn.history['val_loss'], linestyle='--', label='With BN')

plt.xlabel("Epochs")
plt.ylabel("Validation Loss")
plt.title("Figure 2 – Training Stability (BN vs No BN)")
plt.legend()

plt.savefig("figure2_bn_loss.png", dpi=300)
plt.show()
```

### Convergence Speed Plot

```python
epochs = range(len(history_bn.history['accuracy']))

plt.figure(figsize=(8,5))

plt.plot(epochs, history_no_bn.history['accuracy'], label='No BN')
plt.plot(epochs, history_bn.history['accuracy'], linestyle='--', label='With BN')

plt.xlabel("Epochs")
plt.ylabel("Training Accuracy")
plt.title("Figure 3 – Convergence Speed Comparison")
plt.legend()

plt.savefig("figure3_convergence.png", dpi=300)
plt.show()
```

### ADVANCED: BN + Dropout

```python
model_bn_dropout = create_cnn(use_bn=True, use_dropout=True)

history_bn_dropout = model_bn_dropout.fit(
    X_train, y_train,
    validation_split=0.2,
    epochs=8,
    batch_size=128,
    verbose=1
)
```

### Compare All Models

```python
plt.figure(figsize=(8,5))

plt.plot(history_no_bn.history['val_accuracy'], label='No BN')
plt.plot(history_bn.history['val_accuracy'], label='BN')
plt.plot(history_bn_dropout.history['val_accuracy'], linestyle='--', label='BN + Dropout')

plt.xlabel("Epochs")
plt.ylabel("Validation Accuracy")
plt.title("Figure 4 – Model Comparison")
plt.legend()

plt.savefig("figure4_model_compare.png", dpi=300)
plt.show()
```
